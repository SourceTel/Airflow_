```python
import numpy as np
import pandas as pd
import urllib.parse
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
URL = "https://space.skyrocket.de/directories/sat_eo_usa.htm"
```

```python
response = requests.get(URL)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
response.status_code
```

```python
test_url = soup.find(class_ = 'index').find('a')
test_url
```

```python
test_urls = soup.find(class_ = 'index').find_all('a')
test_urls
```

```python
url_list_2 = []

for test in test_urls:
    url_list_2.append(test.get('href'))
```

```python
url_list_2
```

```python
url_list_1 = 'https://space.skyrocket.de/'
```

```python
url_complete = []
for urls in url_list_2:
    url_complete.append(url_list_1 + urls)
```

```python
url_complete
```

```python
results = []

for link in url_complete:
    response = requests.get(link)

    soup = BeautifulSoup(response.content, 'html.parser')

    #Satellite Name
    try:
        sat = soup.find(id = 'satlist').find('td').get_text()
    except:
        sat = 'n/a'
    
    #Satellite Type
    try:
        stype = soup.find(id = 'sdtyp').get_text()
    except:
        stype = 'n/a'

    
    #Satellite Operator
    try:
        sop = soup.find(id = 'sdope').get_text()
    except:
        sop = 'n/a'

    
    #Satellite Contractor
    try:
        scont = soup.find(id = 'sdcon').get_text()
    except:
        scont = 'n/a'

    
    #Satellite Equipment
    try:
        sequip = soup.find(id = 'sdequ').get_text()
    except:
        sequip = 'n/a'

    
    #Satellite Platform
    try:
        sbus = soup.find(id = 'sdcnf').get_text()
    except:
        sbus = 'n/a'

    
    #Satellite Launch Date
    try:
        ld = soup.find(id = 'satlist').find_all('td')[2].get_text()
    except:
        ld = 'n/a'

    
    #Satellite launch Vehicle
    try:
        lv = soup.find(id = 'satlist').find('a').get_text()
    except:
        lv = 'n/a'

    output = {'Satellite': sat, 'Satellite Type': stype, 'Satellite Operator': sop, 'Satellite Contractor': scont, 'Equipment': sequip, 
              'Satellite Platform': sbus, 'Launch Date': ld, 'Launch Vehicle': lv, 'Source': link}
    
    #Results
    results.append(output)
```

```python
df = pd.DataFrame(results)
```
